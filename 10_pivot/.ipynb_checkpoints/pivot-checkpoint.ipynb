```python
df = pd.read_csv('weather2.csv')
df
```

```python
#pivot_table works for the duplicate data
# it takes the mean valuemd
df.pivot_table(index='city', columns='date', margins=True)
```

```python
# it will generate error
df.pivot(index='city', columns='date')
```

```python
#aggfunc is the function which decides what to do with duplicate values
df.pivot_table(index='city', columns='date', aggfunc='sum')
```
