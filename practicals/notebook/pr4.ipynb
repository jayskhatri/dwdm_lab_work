# **Association Rule Mining in Python**

Association rule mining finds interesting associations and relationships among large sets of data items. This rule shows how frequently a itemset occurs in a transaction 

# Measures of Effectiveness of the Rule:
**Support:** Support means how much historical data supports your rule 

**Confidence:** Confidence means how confident are we that the rule holds.

**Lift:** Lift is the ratio Confidence is to Support. If the lift is < 1 then A and B are negatively correlated else positively correlated and if it is equal to 1 it is not correlated.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from apyori import apriori
```

```python
df=pd.read_csv("Dataset/store_data.csv",header=None)
```

```python
df.head()
```

# Implementation of association rule mining using Apriori Algorithm

```python
records=[]
for i in range(0,7501):
    records.append([str(df.values[i,j]) for j in range(0,20)])
```

```python
asso_rules=apriori(records,min_support=0.002,min_confidence=0.2,min_lift=3,min_length=2)
```

```python
asso_result=list(asso_rules)
```

```python
len(asso_result)
```

```python
asso_result[0]
```

# Followings are the rules which are generated using Apriori Algorithm

```python
for item in asso_result:
    pair=item[0]
    items=[x for x in pair]
    print("Rule for: " +items[0]+" -> "+items[1])
    
    print("Support: " +str(item[1]))
    
    print("Confidence"+str(item[2][0][2]))
    print("Lift: " +str(item[2][0][3]))
    print("*****************************")
    
```
