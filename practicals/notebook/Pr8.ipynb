```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from apyori import apriori
```

```python
df=pd.read_csv("Dataset/store_data.csv",header=None)
```

```python
df.head()
```

```python
records=[]
for i in range(0,7501):
    records.append([str(df.values[i,j]) for j in range(0,20)])
```

```python
asso_rules=apriori(records,min_support=0.002,min_confidence=0.2,min_lift=3,min_length=2)
```

```python
asso_result=list(asso_rules)
```

```python
len(asso_result)
```

```python
asso_result[0]
```

```python
for item in asso_result:
    pair=item[0]
    items=[x for x in pair]
    print("Purchase: " +items[0]+" ,Recommending: "+items[1])
    print("-----------")
```
