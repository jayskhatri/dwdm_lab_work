```python
import pandas as pd
df = pd.read_csv('weather_data.csv')
#df = pd.read_csv('weather_data.csv',parse_dates=['day'])
df
```

```python
df.set_index('day',inplace=False)
df
```

```python
df['temperature'].mean()
```

```python
df1 = df.fillna({
    'temperature': df['temperature'].mean(),
    'windspeed': 8000,
    'event': 'No Event'
})
df1.set_index('day',inplace=True)
df1
```

```python
new_df = df.fillna(method='ffill',limit=1)
new_df
new_df = new_df.interpolate()
new_df.set_index('day',inplace=True)
new_df
```

```python
new = df.dropna(thresh=2)#thresh=2   how='all'
new
```
