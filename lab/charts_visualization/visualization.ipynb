```python
#import statements
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('stocknews/upload_DJIA_table.csv')
df.head()
```

```python
df.columns = ['date','opening','maximum','minimum','closing','volume','adj_close']
plt.scatter(df.date[ : 10],df.maximum[ : 10],marker='o')
plt.show()
```

```python
df.tail().boxplot(by='date', column=['volume'])
```

```python
df['maximum'].plot.kde()
```

```python
sy = pd.read_csv('data/spotify.csv')
sy.columns = ['date','shape_of_you','despacito','something_just_like','humble','unforgettable']
sy.fillna({'despacito': sy.despacito.median(), 'something_just_like': sy.something_just_like.median(),'humble':sy.humble.median(), 'unforgettable':sy.unforgettable.median()},inplace=True)
sy.head()
```
